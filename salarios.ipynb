```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/data_science_salaries.csv')
df
```

```python
# cambiando nombre de datos employee_residence y company_location
```

```python
df
```

```python
dummie1 = pd.get_dummies(df['job_title'])
dummie2 = pd.get_dummies(df['experience_level'])
dummie3 = pd.get_dummies(df['employment_type'])
dummie4 = pd.get_dummies(df['work_models'])
dummie5 = pd.get_dummies(df['employee_residence'], prefix='er_')
dummie6 = pd.get_dummies(df['company_location'], prefix='cl_')
dummie7 = pd.get_dummies(df['company_size'], prefix='cs')

dummies = dummie1, dummie2,dummie3, dummie4, dummie5, dummie6, dummie7
```

```python
#df = df.join(dummies)
df
```

```python
#eliminar columnas de donde se sacaron los dummies
```

```python
df.drop(['job_title'], axis=1, inplace=True)
df.drop(['experience_level'], axis=1, inplace=True)
df.drop(['employment_type'], axis=1, inplace=True)
df.drop(['work_models'], axis=1, inplace=True)
df.drop(['employee_residence'], axis=1, inplace=True)
df.drop(['company_location'], axis=1, inplace=True)
df.drop(['company_size'], axis=1, inplace=True)
df.drop(['salary_currency'], axis=1, inplace=True)
df.drop(['salary'], axis=1, inplace=True)
```

```python
#modelo ML
```

```python
x = df.drop(['salary_in_usd'], axis=1)
y = df['salary_in_usd']

model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

x_ent, x_pru, y_ent, y_pru = train_test_split(x, y, test_size=.2)
```

```python
model.fit(x_ent, y_ent)
```

```python
prediccion = model.predict(x_pru)
```

```python
comparativa = {'prediccion': prediccion, 'valor real': y_pru}
pd.DataFrame(comparativa)
```

```python
model.score(x_ent, y_ent)
```

# **Limpieza de datos**

## Benefits tiene muchos datos Nan (Se remplaza con 0)

```python
df.isna().sum()
```

```python
df.Benefits.fillna(int(0), inplace=True)
```

## Eliminar columnas Notes y Status

```python
df.isna().sum()
```

```python
df.drop(columns=['Notes','Status'], axis=1, inplace=True)
```

## covertir objetos a Float

```python
df.info()
```

```python
filas_a_eliminar = []

# Iterar sobre cada columna y encontrar las filas que contienen 'Not Provided'
for columna in df.columns:
    filas_a_eliminar.extend(df[df[columna] == 'Not Provided'].index)

# Eliminar las filas duplicadas de la lista
filas_a_eliminar = list(set(filas_a_eliminar))

# Eliminar las filas encontradas
df = df.drop(filas_a_eliminar)

# Mostrar el DataFrame resultante
df
```

```python
columnas_a_convertir = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']

# Convertir las columnas seleccionadas a tipo float
df[columnas_a_convertir] = df[columnas_a_convertir].astype(float)
```

### Pre Modelado

```python
df_model = df.copy()
```

```python
# se elimina EmployeeName y Agency por se incecesarios

df_model.drop(columns=['EmployeeName','Agency'], inplace=True)
```

```python
df_model
```

```python
from sklearn.preprocessing import OneHotEncoder

#Categóricas
cat = df_model.select_dtypes('O')

#Instanciamos
ohe = OneHotEncoder(sparse_output = False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

```python
df_model.drop(columns=['BasePay', 'OvertimePay', 'TotalPay','JobTitle', 'OtherPay'], inplace=True)
```

```python
df_model = df_model.join(cat_ohe)
```

```python
df_model = df_model.dropna()
```

```python
df_model.isna().any()
```

```python
x = df_model.drop(columns=['TotalPayBenefits'])
y = df_model.TotalPayBenefits
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
prediccion = model.predict(x_test)
```

```python
prediccion
```

```python
y_test
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, prediccion)
```
